```python
# Contagem de eventos por MÊS
contagem_mensal = estelionato_23.groupby(['Ano', 'Mês']).size().reset_index(name='quantidade')

print("Contagem de Eventos Mensais:")
print(contagem_mensal)
```

### Contagem de Eventos por Horário


```python
# Contar os eventos por HORÁRIO
contagem_horario = estelionato_23['Hora'].value_counts().reset_index()
contagem_horario.columns = ['Hora', 'quantidade']

# Ordenar os resultados por horário
contagem_horario.sort_values('Hora', inplace=True)

# Exibir resultado
print("Contagem de Eventos por Horário:")
print(contagem_horario)
```

## Análise de Correlação
#### Para entender as relações entre diferentes variáveis (por exemplo, a relação entre hora e quantidade de estelionatos, ou entre mês e quantidade de estelionatos), podemos calcular a correlação entre as variáveis numéricas.

## Concatenar os Datasets para Análises Comparativas
Para facilitar as comparações, podemos combinar os dois datasets em um único, adicionando uma coluna para identificar o ano.

```python
# Adicionar coluna para identificar o ano
estelionato_23['Ano'] = 2023
estelionato_24['Ano'] = 2024

# Concatenar os datasets
estelionato_completo = pd.concat([estelionato_23, estelionato_24])

# Verificar as primeiras linhas do dataset combinado
print(estelionato_completo.head())
```

## Comparar a Quantidade de Estelionatos por Ano
Uma análise inicial pode ser comparar o número total de estelionatos em 2023 e 2024.

```python
# Contagem total de estelionatos por ano
contagem_ano = estelionato_completo.groupby('Ano').size()
print(contagem_ano)
```

```python
contagem_ano.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'orange'])

plt.title('Número de Estelionatos por Ano (2023 vs 2024)')
plt.xlabel('Ano')
plt.ylabel('Número de Estelionatos')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Análise de Tendências Mensais
Uma análise interessante seria comparar a distribuição dos estelionatos ao longo dos meses de 2023 e 2024.

```python
# Contagem mensal por ano
contagem_mensal_2 = estelionato_completo.groupby(['Ano', 'Mês']).size().reset_index(name='quantidade')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Mês', y='quantidade', hue='Ano', data=contagem_mensal_2, palette='muted')
plt.title('Estelionatos por Mês (2023 vs 2024)')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Estelionatos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Análise de Estelionatos por Dia da Semana

### Teste de Diferenças Entre 2023 e 2024
Você pode realizar testes estatísticos para verificar se há diferenças significativas nas médias de estelionatos entre os dois anos (2023 vs 2024). O teste t pode ser útil.

### Adicionando a Coluna "Quantidade"
Criar a coluna "quantidade" a partir da contagem de registros de estelionatos para cada ano, mês. 

```python
# Contar o número de estelionatos por Ano
contagem_ano = estelionato_completo.groupby('Ano').size().reset_index(name='quantidade')

# Visualizar a contagem
print(contagem_ano)
```

```python
# Contagem total de estelionatos por ano
contagem_ano = estelionato_completo.groupby('Ano').size()

# Verificar os valores totais de estelionatos em 2023 e 2024
total_2023 = contagem_ano[2023]
total_2024 = contagem_ano[2024]

print(f"Total de estelionatos em 2023: {total_2023}")
print(f"Total de estelionatos em 2024: {total_2024}")
```

```python
# Calcular a variação percentual
variacao_percentual = ((total_2024 - total_2023) / total_2023) * 100

# Exibir o resultado
if variacao_percentual > 0:
    print(f"Houve um aumento de {variacao_percentual:.2f}% no número de estelionatos de 2023 para 2024.")
else:
    print(f"Houve uma diminuição de {abs(variacao_percentual):.2f}% no número de estelionatos de 2023 para 2024.")
```
